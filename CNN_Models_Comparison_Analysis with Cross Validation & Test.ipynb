## 5. K-Fold Cross-Validation Analysis

### 5.1 Overview

K-Fold Cross-Validation provides a robust evaluation method by training and validating models on different data splits. This section presents comprehensive K-Fold results for all eight models using **StratifiedKFold (5 folds)** to ensure balanced class distribution across folds.

**Methodology:**
- **StratifiedKFold (k=5)**: Ensures each fold maintains the same class distribution as the full dataset
- **Data Split Strategy**: Combined train (80%) + validation (10%) = 90% of total data used for K-Fold
- **Test Set**: 10% of total data reserved and NOT used in K-Fold (reserved for final evaluation)
- **K-Fold Internal Split**: Each fold uses 80% train / 20% validation of the combined 90% data
- **Metrics**: Average accuracy, standard deviation, min/max accuracy, and per-fold accuracies

```python
# ============================================================================
# GRAPH 10: Generalization Gap Analysis (SEPARATE)
# ============================================================================
fig, ax = plt.subplots(figsize=(14, 8))

# Color coding: green for good generalization, yellow for moderate, red for poor
colors_gen = []
for gap in generalization_gap:
    if abs(gap) < 2:
        colors_gen.append('#2E86AB')  # Excellent generalization
    elif abs(gap) < 5:
        colors_gen.append('#F18F01')  # Moderate gap
    else:
        colors_gen.append('#C73E1D')  # Poor generalization

bars = ax.barh(models, generalization_gap, color=colors_gen, alpha=0.8, edgecolor='black', linewidth=1.5)

# Add value labels
for i, (bar, gap) in enumerate(zip(bars, generalization_gap)):
    label = f'{gap:.2f}%'
    ax.text(gap + (1 if gap >= 0 else -1), i, label, va='center', 
            fontsize=11, fontweight='bold', ha='left' if gap >= 0 else 'right')

# Add zero line
ax.axvline(x=0, color='black', linestyle='--', linewidth=2, alpha=0.5, label='Perfect Generalization')

ax.set_xlabel('Generalization Gap (%) = Validation Accuracy - Test Accuracy', 
              fontsize=12, fontweight='bold')
ax.set_title('Generalization Analysis: Validation vs Test Accuracy Gap', 
             fontsize=16, fontweight='bold', pad=20)
ax.grid(axis='x', alpha=0.3, linestyle='--')

# Add legend
legend_elements = [
    Patch(facecolor='#2E86AB', label='Excellent Generalization (<2% gap)'),
    Patch(facecolor='#F18F01', label='Moderate Gap (2-5%)'),
    Patch(facecolor='#C73E1D', label='Poor Generalization (>5% gap)')
]
ax.legend(handles=legend_elements, loc='lower right', fontsize=11)

plt.tight_layout()
plt.show()
print("üìä Graph 10: Generalization Gap Analysis")
```
